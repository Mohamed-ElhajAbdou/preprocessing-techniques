```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
image = plt.imread('pic.jpg')
plt.imshow(image)
```

```python
image_adjusting_the_brightness=tf.image.adjust_brightness(image,0.4)
```

```python
image_adjusting_the_brightness_1=tf.image.adjust_brightness(image,0.2)
image_adjusting_the_brightness_2=tf.image.adjust_brightness(image,0.6)
```

```python
plt.imshow(image_adjusting_the_brightness_1)
```

```python
plt.imshow(image_adjusting_the_brightness_2)
```

```python
image_adjust_contrast=tf.image.adjust_contrast(image,0.5)
```

```python
plt.imshow(image_adjust_contrast)
```

```python
image_gamma=tf.image.adjust_gamma(image,gamma=1,gain=1)
```

```python
plt.imshow(image_gamma)
```

```python
hue=tf.image.adjust_hue(image,0.9,name=None)
```

```python
plt.imshow(hue)
```

```python
adjust_saturation=tf.image.adjust_saturation(image,9,name=None)
```

```python
plt.imshow(adjust_saturation)
```

```python
crop_center=tf.image.central_crop(image,1)
```

```python
plt.imshow(crop_center)
```

```python
crop_2=tf.image.crop_to_bounding_box(image,20,10,300,400)
```

```python
plt.imshow(crop_2)
```

```python
flib_left_right=tf.image.flip_left_right(image)
plt.imshow(flib_left_right)
```

```python
flib_up_down=tf.image.flip_up_down(image)
plt.imshow(flib_up_down)
```

```python
gray_scale=tf.image.rgb_to_grayscale(image,name=None)
print(gray_scale.shape)
```

```python
gray_scale=tf.squeeze(gray_scale)
print(gray_scale.shape)
```

```python
plt.imshow(gray_scale,cmap='gray')
```

```python
rotate=tf.image.rot90(image,k=4,name=None)
```

```python
plt.imshow(rotate)
```

```python
Transpose=tf.image.transpose(image,name=None)
plt.imshow(Transpose)
```
